## Exploratory Data Analysis (EDA)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import missingno as msgn
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('walmart.csv')
df.head()
```

```python
for col in df.columns:
    print(col,': ',df[col].nunique())
```

```python
df.Date = pd.to_datetime(df.Date)
df['Month'] = df.Date.dt.month_name()
df['Day'] = df.Date.dt.day_name()
df['Year'] = df.Date.dt.year
```

```python
df.head()
```

## Let's see the Sales in Year, Month and day of the week

```python
plt.figure(figsize=(8,6))
plt.xlabel("Year",fontsize=16)
plt.ylabel("Count",fontsize=16)
plt.yticks(fontsize=14)
plt.xticks(fontsize=14)
sns.countplot(x=df['Year'],order=df['Year'].value_counts().index,palette='CMRmap_r')
plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.xlabel("Month",fontsize=16)
plt.ylabel("Count",fontsize=16)
plt.yticks(fontsize=14)
plt.xticks(fontsize=14,rotation=45)
sns.countplot(df['Month'],order=df['Month'].value_counts().index,palette='gnuplot')
```

```python
plt.figure(figsize=(9,7))
plt.xlabel("Day",fontsize=16)
plt.ylabel("Count",fontsize=16)
plt.yticks(fontsize=14)
plt.xticks(fontsize=14,rotation=45)
sns.countplot(df['Day'],order=df['Day'].value_counts().index,palette='rainbow')
```

```python
df.Store.value_counts()
```

```python
df_temp = df.groupby('Store')['Weekly_Sales'].mean().round(2).sort_values(ascending=False).head(5)
```

```python
sns.barplot(x=df_temp.index,y=df_temp.values,order=df_temp.index,palette='cool_r')
```

```python
df.groupby('Year')['Weekly_Sales'].mean().round(2)
```

```python
plt.figure(figsize=(8,7))
plt.pie(df.groupby('Year')['Weekly_Sales'].sum(),labels=df['Year'].unique(),autopct='%1.2f%%',normalize=True,textprops={'fontsize':15},colors=['Green','Blue','Magenta'])
plt.title('Yearly Sales',fontsize=20)
plt.show()
```

```python
df_temp2 = df.groupby('Month')['Weekly_Sales'].sum().sort_values(ascending=False).reset_index()
df_temp2.head(15)
```

```python
plt.figure(figsize=(10,15))
plt.pie(df_temp2['Weekly_Sales'],labels=df_temp2['Month'],autopct='%1.2f%%',normalize=True,textprops={'fontsize':12})
plt.title('Monthly Sales',fontsize=20)
plt.show()
```

```python
df_temp3= df.groupby('Day')['Weekly_Sales'].sum().reset_index()
df_temp3.head(10)
```

```python
plt.figure(figsize=(10,13))
plt.pie(df_temp3['Weekly_Sales'],labels=df_temp3['Day'],autopct='%1.2f%%',normalize=True,textprops={'fontsize':14})
plt.title('Weekly Sales',fontsize=20)
plt.show()
```

## Weekly Sales by Year

```python
df_2010 = df[df["Year"] ==  2010]
df_2011 = df[df["Year"] == 2011]
df_2012 = df[df["Year"] == 2012]
```

```python
plt.figure(figsize=(12,7))
sns.set_style('darkgrid')
plt.title('Weekly Sales by Month ,2010',fontdict={'fontsize':20})
df_2010.groupby('Month')['Weekly_Sales'].mean().plot(linewidth=2,style='--o')
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.set_style('darkgrid')
plt.title('Weekly Sales by Month ,2011',fontdict={'fontsize':20})
df_2011.groupby('Month')['Weekly_Sales'].mean().plot(linewidth=2,style='--go')
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.set_style('darkgrid')
plt.title('Weekly Sales by Month ,2012',fontdict={'fontsize':20})
df_2012.groupby('Month')['Weekly_Sales'].mean().plot(linewidth=2,style='--ro',)
plt.show()
```
