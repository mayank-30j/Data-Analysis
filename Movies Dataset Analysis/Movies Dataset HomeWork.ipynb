```python
import pandas as pd
import codecs
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

```python
with codecs.open("Movies Dataset.csv", "r",encoding='utf-8', errors='ignore') as file_dat:
     movies = pd.read_csv('Movies Dataset.csv')
```

## Data Cleaning and Basic level EDA

```python
movies.head()
```

```python
movies.shape # 608 rows and 18 cols 
```

```python
# Removing space from the column name
movies.columns = movies.columns.str.replace(' ','')

# Replacing the brackets 
movies.columns = movies.columns.str.replace('(','').str.replace(')','')

# Removing $ sign 
movies.columns = movies.columns.str.replace('$','')

# Replacing mill with Million
movies.columns = movies.columns.str.replace('mill','Million')

# Removing % sign
movies.columns = movies.columns.str.replace('%','')

# Renaming some the columns
movies.rename(columns={'DayofWeek':'DayOfRelease','Overseas':'OverSeas_per','Runtimemin':'RunTime','Profit':'Profit_per'},inplace=True)
```

```python
# Column names are finally cleaned
movies.columns
```

```python
movies['Director'].value_counts().head(10)
```

```python
for col in ['AdjustedGrossMillion','ProfitMillion','OverseasMillion','GrossMillion']: 
    movies[col] = [float(i.replace(',','')) for i in movies[col]]
```

```python
movies.head()
```

```python
print('This is an Averge of all the amounts and not total sum')
print()

for name in movies['Director'].unique():
    print('Director Name: ',name)
    
    budget = movies[movies['Director'] == f'{name}'].BudgetMillion.mean()
    print(f'Budget :  ${round(budget,2)} million')
    
    gross = movies[movies['Director'] == f'{name}'].AdjustedGrossMillion.mean()
    print(f'Gross :  ${round(gross,2)} million') 

    profit = movies[movies['Director'] == f'{name}'].ProfitMillion.mean()
    print(f'Profit:  ${round(profit,2)} million')
    
    print()
```

```python
movies['ReleaseDate'] = pd.to_datetime(movies['ReleaseDate'])
```
